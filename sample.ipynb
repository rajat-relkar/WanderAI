```python
def remove_duplicates(nums):
    k = 1 
    
    for i in range(1, len(nums)):
        if nums[i] != nums[k - 1]:
            nums[k] = nums[i]
            k += 1

    return k

nums_1 = [1, 1, 2]
print(remove_duplicates(nums_1))
```
